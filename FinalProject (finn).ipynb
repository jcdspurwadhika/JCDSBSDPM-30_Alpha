## 2.1 Latar Belakang Bisnis Telemarketing Bank

Deposito merupakan produk simpanan berjangka di bank, di mana nasabah menyimpan dana untuk periode tertentu dengan tingkat bunga tetap. Berbeda dengan tabungan biasa, dana deposito tidak dapat ditarik sewaktu-waktu tanpa penalti. salah satu sumber utama dana pihak ketiga bagi bank adalah produk term deposit (deposito berjangka) karena dana ini berperan penting dalam menjaga likuiditas, mendukung penyaluran kredit, serta menjaga stabilitas operasional bank. Oleh karena itu, keberhasilan penjualan produk deposito memiliki dampak langsung terhadap kinerja finansial bank.

### Dampak Deposito terhadap Bisnis Bank
- **Sumber Dana Stabil:** Deposito memberikan likuiditas jangka menengah hingga panjang yang lebih terjamin dibandingkan tabungan.  
- **Pendapatan Bunga:** Dana deposito dapat disalurkan ke kredit dengan bunga lebih tinggi, sehingga menghasilkan margin keuntungan bagi bank.  
- **Kepercayaan Nasabah:** Tingginya jumlah deposito mencerminkan tingkat kepercayaan masyarakat terhadap bank.  
- **Strategi Pertumbuhan:** Produk deposito sering digunakan untuk menarik nasabah baru sekaligus memperkuat basis dana pihak ketiga.

Salah satu kanal pemasaran yang umum digunakan untuk menawarkan produk deposito adalah telemarketing. Melalui telemarketing, bank dapat menjangkau nasabah secara langsung dan personal. 

### Alasan Telemarketing Digunakan
- **Targeted Marketing:** Telemarketing memungkinkan bank menghubungi langsung calon nasabah yang sudah ada dalam database.  
- **Efisiensi Biaya:** Lebih hemat dibandingkan membuka cabang baru atau melakukan kampanye massal.  
- **Pendekatan Personal:** Memberikan kesempatan untuk menjelaskan detail produk secara langsung dan menjawab pertanyaan nasabah.  
- **Cepat dan Terukur:** Hasil kampanye dapat segera diukur melalui jumlah panggilan, respon, dan tingkat konversi.

### Tantangan dalam Telemarketing
- **Intrusif:** Banyak nasabah merasa terganggu dengan panggilan yang tidak diinginkan.  
- **Efektivitas Rendah:** Tingkat konversi biasanya kecil karena mayoritas nasabah menolak atau tidak tertarik.  
- **Biaya Operasional:** Meski lebih murah dari kampanye besar, tetap membutuhkan tenaga kerja dan waktu yang signifikan.  
- **Reputasi Bank:** Panggilan berlebihan dapat menurunkan citra bank karena dianggap mengganggu privasi nasabah.  
- **Data Imbalance:** Hanya sebagian kecil nasabah yang benar-benar tertarik, sehingga diperlukan model prediksi untuk meningkatkan efisiensi targeting.

##  2.5 Metric Evaluation dan Konsekuensi Kesalahan Prediksi

Dalam konteks kampanye telemarketing deposito berjangka, kesalahan prediksi model memiliki implikasi bisnis yang berbeda.

- **Type I Error (False Positive)**  
  Terjadi ketika model memprediksi nasabah sebagai potensial, namun nasabah tersebut tidak melakukan pembukaan deposito. Konsekuensi dari kesalahan ini adalah pemborosan biaya operasional, waktu tenaga telemarketing, serta potensi penurunan pengalaman nasabah akibat panggilan yang tidak relevan.

- **Type II Error (False Negative)**  
  Terjadi ketika model gagal mengidentifikasi nasabah yang sebenarnya memiliki potensi untuk berlangganan deposito. Konsekuensinya adalah hilangnya peluang bisnis, namun tanpa menimbulkan biaya operasional langsung karena panggilan tidak dilakukan.

Berdasarkan perbedaan konsekuensi tersebut, kesalahan **false positive** dinilai lebih kritis dibandingkan **false negative**, karena secara langsung merepresentasikan pemborosan sumber daya telemarketing. Oleh karena itu, model perlu difokuskan pada peningkatan efisiensi panggilan, tanpa mengorbankan terlalu banyak peluang potensial. 

Dalam konteks kampanye telemarketing, biaya operasional per panggilan bukanlah angka teoretis. Laporan industri call center menunjukkan bahwa biaya rata-rata per panggilan dapat mencapai sekitar *USD 7.16 per panggilan*, setelah memperhitungkan komponen biaya seperti gaji agen, infrastruktur telekomunikasi, serta perangkat lunak pendukung seperti *Customer Relationship Management* (CRM) dan sistem dialer (Voiso, 2023; Abacus BPO, 2023). Hal ini menunjukkan bahwa setiap panggilan yang dilakukan dalam kampanye telemarketing memiliki biaya operasional nyata bagi perusahaan. Sebagai ilustrasi sederhana, apabila sebuah kampanye telemarketing menargetkan *1.000 nasabah* dengan biaya rata-rata sekitar *USD 7 per panggilan*, maka total biaya yang dikeluarkan hanya untuk aktivitas panggilan dapat mencapai sekitar *USD 7.000*. Biaya ini akan meningkat secara proporsional seiring bertambahnya jumlah panggilan yang tidak menghasilkan konversi, khususnya akibat *false positive*, yaitu ketika model memprediksi nasabah sebagai prospek potensial namun pada kenyataannya tidak melakukan konversi.

Dalam konteks pemodelan prediksi, *false positive*—yaitu kondisi ketika model memprediksi nasabah sebagai prospek potensial namun pada kenyataannya tidak melakukan konversi—akan menyebabkan bank tetap melakukan panggilan dan menanggung biaya operasional tersebut tanpa menghasilkan pendapatan (Voiso, 2023). Sebaliknya, *false negative* merepresentasikan peluang bisnis yang tidak dimanfaatkan, namun tidak menimbulkan biaya panggilan secara langsung karena kontak tidak dilakukan, meskipun berpotensi mengorbankan peluang konversi (Abacus BPO, 2023).

Selain biaya per panggilan, efektivitas kampanye telemarketing juga sering diukur menggunakan pendekatan *Cost Per Lead* (CPL), yaitu total biaya kampanye dibagi jumlah prospek yang dihasilkan. Sebagai ilustrasi, apabila total biaya telemarketing sebesar *USD 10.000* menghasilkan *500 prospek*, maka biaya per prospek (CPL) adalah *USD 20 per lead*, yang menunjukkan bagaimana biaya kampanye dapat dikaitkan langsung dengan hasil yang diperoleh (Teledata Marketing, 2022). Pendekatan ini memperlihatkan bahwa kesalahan prediksi, khususnya *false positive*, dapat secara langsung meningkatkan biaya kampanye tanpa memberikan nilai bisnis, sementara *false negative* lebih bersifat sebagai *biaya peluang (opportunity cost)*. Oleh karena itu, pemilihan metrik evaluasi model perlu mempertimbangkan trade-off antara efisiensi biaya operasional dan potensi pertumbuhan bisnis (Voiso, 2023; Teledata Marketing, 2022).

Dalam proyek ini, **precision** dipilih sebagai metrik utama karena secara langsung mengukur proporsi panggilan yang benar-benar menghasilkan konversi. **Recall** digunakan sebagai metrik pendukung untuk memastikan model tidak terlalu restriktif dan tetap mampu menangkap peluang yang relevan.

## 3.4 Conversion Rate Analysis
Memahami baseline konversi sangat penting untuk menetapkan standar performa model. Dalam konteks bisnis ini, conversion rate didefinisikan sebagai proporsi nasabah yang membuka produk term deposit dibandingkan dengan seluruh nasabah yang dihubungi melalui telemarketing campaign.

````python
# Visualisasi Class Imbalance
ax = sns.countplot(x='y', data=df, palette=['#e74c3c', '#2ecc71'])
plt.title('Distribusi Target Variable (y): Class Imbalance', fontsize=14)
plt.xlabel('Langganan Deposito?', fontsize=12)
plt.ylabel('Jumlah Nasabah', fontsize=12)

# Menambahkan label persentase
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(df))
    x = p.get_x() + p.get_width() / 2 - 0.05
    y = p.get_height()
    ax.annotate(percentage, (x, y), size = 12, va='bottom')

plt.show()

# Menghitung Conversion Rate Alami
conv_counts = df['y'].value_counts()
conv_rate = df['y'].value_counts(normalize=True) * 100

print(f"Total Nasabah yang Dihubungi: {len(df)}")
print(f"Total Sukses Konversi: {conv_counts['yes']}")
print(f"Baseline Conversion Rate: {conv_rate['yes']:.2f}%")
````

**Implikasi Bisnis**: Dari total 45.211 nasabah yang dihubungi, hanya 5.289 nasabah yang akhirnya membuka deposito berjangka. Hal ini menghasilkan baseline conversion rate sebesar 11,70%, yang berarti bahwa hampir 9 dari 10 panggilan telemarketing tidak menghasilkan konversi. Model kita harus mampu melakukan "LIFT" terhadap angka 11.7% ini pada kelompok prioritas.

Namun, dalam kampanye telemarketing bank, angka ini **tidak otomatis menunjukkan keberhasilan
operasional**. Setiap panggilan telemarketing memiliki biaya (tenaga kerja, waktu, infrastruktur)
dan kapasitas yang terbatas. Ketika mayoritas panggilan (±88–89%) tidak menghasilkan konversi,
bank menghadapi dilema efisiensi: banyak sumber daya yang terbuang untuk sedikit hasil.

Secara bisnis, *conversion rate* sekitar 10–12% dianggap **normal secara konteks industri telemarketing,
namun potensi inefisiensi tinggi tetap menjadi masalah** ketika biaya dan kapasitas panggilan tidak unlimited.
Oleh karena itu, tujuan utama pengembangan model prediktif bukanlah meningkatkan konversi secara keseluruhan dengan menghubungi semua nasabah, melainkan meningkatkan kualitas target panggilan. Model diharapkan mampu menghasilkan lift, yaitu peningkatan tingkat konversi dibandingkan baseline 11,70%, pada kelompok nasabah yang diprioritaskan. Dengan memfokuskan panggilan hanya pada nasabah dengan probabilitas tertinggi, bank dapat mencapai tingkat konversi yang lebih tinggi dengan jumlah panggilan yang lebih sedikit, sehingga meningkatkan efisiensi dan efektivitas kampanye telemarketing.

# 4. Data Quality Check & Cleaning

Tahap ini bertujuan untuk memastikan integritas data (data integrity) dan kesiapan fitur sebelum memasuki tahap pemodelan. Kami melakukan audit terhadap anomali, redundansi, dan distribusi nilai yang tidak lengkap.

## 4.1 Normalisasi & Integritas Data
Data string seringkali memiliki variasi format (spasi tambahan atau perbedaan kapitalisasi) yang dapat dianggap sebagai kategori berbeda oleh model. Kami melakukan normalisasi standar untuk menjamin konsistensi kategori. Selain itu, pengecekan duplikasi dilakukan untuk menghindari *over-representation* pada profil nasabah tertentu yang dapat menyebabkan bias.


````python
# Langkah 1: Normalisasi kategori string
cat_cols = ['job','marital','education','default','housing','loan','contact','month','poutcome']
for c in cat_cols:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip().str.lower()

# Preview hasil normalisasi
df[cat_cols].head()
````

Pertama kita melakukan normalisasi pada kolom kategorikal dengan mengubah semua nilai menjadi huruf kecil dan menghapus spasi berlebih. Hal ini memastikan kategori yang sama tidak dianggap berbeda hanya karena perbedaan penulisan, misalnya Admin. dan admin akan digabung menjadi satu kategori yang konsisten.

````python
# Langkah 2: Cek duplikasi
dup_count = df.duplicated().sum()
print("Jumlah duplikasi:", dup_count)

# Hapus duplikasi jika ada
if dup_count > 0:
    df = df.drop_duplicates()

# Preview dataset setelah drop duplikasi
df.head()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# Pastikan y dalam bentuk string
# ===============================
df['y'] = df['y'].astype(str)

# Hitung distribusi
target_count = df['y'].value_counts()
target_prop = df['y'].value_counts(normalize=True)

# Urutan konsisten
target_count = target_count.reindex(['no', 'yes'])
target_prop = target_prop.reindex(['no', 'yes'])

# Label rapi
label_map = {'no': 'No', 'yes': 'Yes'}
target_count.index = target_count.index.map(label_map)
target_prop.index = target_prop.index.map(label_map)

# ===============================
# Warna Konsisten
# ===============================
colors = ['firebrick', 'steelblue']  # No = merah, Yes = biru

# ===============================
# Plot
# ===============================
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# ---- BAR CHART ----
sns.barplot(
    x=target_count.index,
    y=target_count.values,
    ax=axes[0],
    palette=colors
)

axes[0].set_title("Distribusi Target (Total Nasabah)")
axes[0].set_xlabel("Status Berlangganan")
axes[0].set_ylabel("Jumlah Nasabah")

max_val = target_count.max()
axes[0].set_ylim(0, max_val * 1.15)

for i, v in enumerate(target_count.values):
    axes[0].text(
        i,
        v + max_val * 0.03,
        f"{v:,}",
        ha='center',
        va='bottom',
        fontweight='bold'
    )

# ---- PIE CHART ----
axes[1].pie(
    target_prop.values,
    labels=target_prop.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors
)

axes[1].set_title("Baseline Conversion Rate")

plt.tight_layout()
plt.show()
plt.close()
````

## 5.2 Segmentasi Demografis – Conversion Rate Berdasarkan Job
Analisis segmentasi berdasarkan **jenis pekerjaan (job)** bertujuan untuk memahami efektivitas kampanye telemarketing dari dua sudut pandang utama, yaitu **tingkat konversi** dan **skala populasi nasabah**. Oleh karena itu, analisis ini disajikan dalam dua visualisasi yang saling melengkapi: *conversion rate* dan *distribusi jumlah nasabah*.

Pendekatan ini penting karena **conversion rate yang tinggi tidak selalu berarti kontribusi bisnis terbesar**, sementara segmen dengan populasi besar belum tentu efisien secara konversi.

````python
# ================================
# Conversion Rate Berdasarkan Job
# ================================

job_conv = (
    df.groupby('job')['y']
    .apply(lambda x: (x == 'yes').mean())
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,4))
sns.barplot(
    x=job_conv.index,
    y=job_conv.values,
    color='steelblue',
    legend=False
)

plt.xticks(rotation=45)
plt.title("Conversion Rate berdasarkan Pekerjaan")
plt.ylabel("Conversion Rate")
plt.xlabel("Job")

for i, v in enumerate(job_conv.values):
    plt.text(i, v + 0.002, f"{v:.2%}", ha='center', fontsize=8)

plt.show()

# ================================
# Distribusi Jumlah Nasabah (Jumlah + Persentase)
# ================================

job_count = df['job'].value_counts()
job_count_sorted = job_count.reindex(job_conv.index)

total_data = job_count_sorted.sum()  # total nasabah

plt.figure(figsize=(10,4))
sns.barplot(
    x=job_count_sorted.index,
    y=job_count_sorted.values,
    color='steelblue'
)

plt.xticks(rotation=45)
plt.title("Distribusi Jumlah Nasabah berdasarkan Pekerjaan\n(Urutan sesuai Conversion Rate)")
plt.ylabel("Jumlah Nasabah")
plt.xlabel("Job")

# Label jumlah + persentase
for i, v in enumerate(job_count_sorted.values):
    percent = v / total_data * 100
    plt.text(
        i,
        v + (0.01 * v),
        f"{v:,} ({percent:.1f}%)",
        ha='center',
        fontsize=8
    )

plt.tight_layout()
plt.show()
````

## 5.5 Intensitas Kontak (Campaign)

Analisis ini bertujuan untuk memahami bagaimana **jumlah kontak yang dilakukan kepada nasabah dalam satu kampanye** memengaruhi peluang konversi pembukaan term deposit. Variabel `campaign` merepresentasikan intensitas usaha telemarketing yang dilakukan terhadap satu nasabah selama periode kampanye berjalan.

Untuk meningkatkan keterbacaan dan interpretasi bisnis, variabel `campaign` dikelompokkan ke dalam beberapa kategori intensitas kontak (*campaign bin*), yaitu:
- **1**: kontak pertama
- **2–3**: follow-up awal
- **4–5**: follow-up intensif
- **6–10**: over-contacting
- **>10**: extreme contacting

````python
# Bin jumlah kontak campaign
# ================================

df['campaign_bin'] = pd.cut(
    df['campaign'],
    bins=[0, 1, 3, 5, 10, 50],
    labels=['1', '2–3', '4–5', '6–10', '>10']
)

# Mapping deskripsi bisnis
campaign_desc = {
    '1': '1: kontak pertama',
    '2–3': '2–3: follow-up awal',
    '4–5': '4–5: follow-up intensif',
    '6–10': '6–10: over-contacting',
    '>10': '>10: extreme contacting'
}

# ================================
# Conversion Rate berdasarkan campaign_bin
# ================================

campaign_bin_conv = (
    df.groupby('campaign_bin')['y']   # asumsi y = 0/1
   .apply(lambda x: (x == 'yes').mean())
    .sort_values(ascending=False)     # ascending agar bagus untuk horizontal
)

# Ganti label dengan deskripsi
conv_index_desc = [campaign_desc[i] for i in campaign_bin_conv.index]

plt.figure(figsize=(8,4))
ax = sns.barplot(
    x=campaign_bin_conv.values,
    y=conv_index_desc,
    color='steelblue'
)

plt.title("Conversion Rate berdasarkan Intensitas Kontak")
plt.xlabel("Conversion Rate")
plt.ylabel("Intensitas Kontak")

# Label persentase
for i, v in enumerate(campaign_bin_conv.values):
    ax.text(
        v + 0.003,
        i,
        f"{v:.2%}",
        va='center',
        fontsize=9
    )

plt.tight_layout()
plt.show()


# ================================
# Distribusi Jumlah Nasabah (Jumlah + Persentase)
# ================================

campaign_bin_count = (
    df['campaign_bin']
    .value_counts()
    .reindex(campaign_bin_conv.index)
)

total_data = campaign_bin_count.sum()
count_index_desc = [campaign_desc[i] for i in campaign_bin_count.index]

plt.figure(figsize=(8,4))
ax = sns.barplot(
    x=campaign_bin_count.values,
    y=count_index_desc,
    color='steelblue'
)

plt.title(
    "Distribusi Jumlah Nasabah berdasarkan Intensitas Kontak\n"
    "(Urutan mengikuti Conversion Rate)"
)
plt.xlabel("Jumlah Nasabah")
plt.ylabel("Intensitas Kontak")

# Label jumlah + persentase
for i, v in enumerate(campaign_bin_count.values):
    percent = v / total_data * 100
    ax.text(
        v + (0.01 * v),
        i,
        f"{v:,} ({percent:.1f}%)",
        va='center',
        fontsize=9
    )

plt.tight_layout()
plt.show()
````

#### Conversion Rate berdasarkan Intensitas Kontak

Visualisasi pertama menunjukkan bahwa **conversion rate menurun secara konsisten seiring meningkatnya jumlah kontak**. Nasabah yang dihubungi **hanya satu kali** memiliki conversion rate tertinggi (sekitar **14–15%**), sementara nasabah yang dihubungi lebih dari 10 kali hanya memiliki conversion rate di kisaran **3–4%**.

Pola ini mengindikasikan adanya fenomena *diminishing returns*, di mana upaya kontak tambahan tidak hanya gagal meningkatkan peluang konversi, tetapi justru berasosiasi dengan penurunan efektivitas.

#### Distribusi Jumlah Nasabah berdasarkan Intensitas Kontak

Visualisasi distribusi menunjukkan bahwa mayoritas nasabah berada pada kategori **1** dan **2–3 kontak**, sedangkan kategori dengan intensitas tinggi (≥6 kontak) hanya mencakup sebagian kecil dari total populasi.

Namun, meskipun kelompok **over-contacted** berukuran relatif kecil, kelompok ini menyerap usaha telemarketing yang jauh lebih besar per nasabah, dengan hasil konversi yang rendah.

#### Insight & Implikasi Bisnis

Beberapa temuan penting dari analisis ini adalah:

1. **Kontak pertama memiliki nilai prediktif yang paling tinggi**, menunjukkan pentingnya kualitas targeting sejak awal.
2. **Penambahan frekuensi kontak tidak meningkatkan konversi**, bahkan cenderung kontraproduktif.
3. Nasabah yang sulit dikonversi sering kali terus dihubungi berulang kali, menyebabkan **inefisiensi alokasi sumber daya**.
4. Strategi telemarketing yang efektif seharusnya **membatasi intensitas kontak** dan mengalihkan fokus pada pemilihan nasabah dengan probabilitas konversi tinggi sejak awal.

Secara keseluruhan, temuan ini memperkuat kebutuhan akan **pendekatan prediktif berbasis data** untuk menentukan *siapa yang perlu dihubungi lebih dahulu*, bukan meningkatkan frekuensi kontak secara indiscriminatif.



## 5.6 Variabel Numerik – Age & Balance

Analisis variabel numerik difokuskan pada **usia (age)** dan **saldo rekening (balance)** karena keduanya merepresentasikan karakteristik demografis dan kapasitas finansial nasabah. Untuk menjaga interpretasi yang stabil dan relevan secara bisnis, visualisasi menggunakan **median** sebagai ukuran pemusatan, bukan rata-rata, mengingat distribusi data yang cenderung skewed dan mengandung nilai ekstrem.

````python
# 1. Ringkasan Statistik Numerik
summary_num = (
    df.groupby('y')
    .agg(
        median_age=('age', 'median'),
        median_balance=('balance', 'median')
    )
    .reset_index()
)

status_labels = ['No Deposit', 'Deposit']

# 2. Setup Figure
# Menambah figsize sedikit agar lebih lega
fig, ax = plt.subplots(1, 2, figsize=(12, 5)) 

# 3. Plot Median Usia
sns.barplot(
    x='y',
    y='median_age',
    data=summary_num,
    ax=ax[0],
    color='steelblue'
)

ax[0].set_title("Median Usia berdasarkan Status", pad=15) # Tambah padding judul
ax[0].set_xlabel("Status Deposit", labelpad=10)
ax[0].set_ylabel("Median Usia")
ax[0].set_xticklabels(status_labels)

# Menaikkan batas Y agar label teks di atas bar punya ruang (1.2 atau 20% ekstra)
ax[0].set_ylim(0, summary_num['median_age'].max() * 1.2)

# Value label dengan padding lebih besar
for container in ax[0].containers:
    ax[0].bar_label(container, fmt='%.1f', padding=8)

# 4. Plot Median Balance
sns.barplot(
    x='y',
    y='median_balance',
    data=summary_num,
    ax=ax[1],
    color='steelblue'
)

ax[1].set_title("Median Saldo berdasarkan Status", pad=15)
ax[1].set_xlabel("Status Deposit", labelpad=10)
ax[1].set_ylabel("Median Saldo (Log Scale)")
ax[1].set_yscale('linear') # Disarankan linear jika angkanya tidak terpaut ribuan kali lipat agar label terbaca jelas

ax[1].set_xticklabels(status_labels)

# Menaikkan batas Y secara signifikan (1.3 atau 30% ekstra) agar label balance tidak mepet atas
ax[1].set_ylim(0, summary_num['median_balance'].max() * 1.3)

# Value label (ribuan) dengan padding lebih besar
for container in ax[1].containers:
    ax[1].bar_label(
        container, 
        labels=[f"{int(v):,}" for v in summary_num['median_balance']], 
        padding=8
    )

# 5. Final Touch
# h_pad dan w_pad memberikan ruang antar subplot
plt.tight_layout(pad=3.0) 
plt.show()
````

## 6.3 Data Leakage Prevention: Exclusion of Post-Call Information

Berdasarkan EDA, variabel `duration` (durasi panggilan terakhir) memiliki korelasi yang sangat kuat dengan target. 
Namun secara operasional:

- Nilai `duration` **baru diketahui setelah panggilan selesai**
- Model digunakan untuk **menentukan nasabah mana yang perlu dihubungi terlebih dahulu**

Artinya, `duration` tidak tersedia pada saat prediksi dilakukan dan termasuk **data leakage**.

Menggunakan variabel ini akan menghasilkan performa model yang tampak sangat tinggi, tetapi **tidak realistis** dan 
**tidak dapat diterapkan dalam dunia nyata**.

Oleh karena itu, variabel `duration`:
- Tetap digunakan dalam EDA (untuk analisis deskriptif)
- **Dihapus dari dataset modeling**



````python
# Drop kolom leakage-prone (duration hanya diketahui setelah call)
if 'duration' in df_model.columns:
    df_model.drop(columns=['duration'], inplace=True)
    print("Kolom 'duration' dihapus karena data leakage.")

# Preview kolom setelah drop
df_model.head()

````

Hasil eksekusi kode menunjukkan bahwa variabel `duration` berhasil dihapus dari dataset pemodelan. Dengan langkah ini, seluruh fitur yang tersisa merepresentasikan informasi yang tersedia sebelum call dilakukan, sehingga model yang dibangun tetap implementable dalam konteks bisnis nyata.

Keputusan ini selaras dengan prinsip *implementable predictive targeting* yang dikemukakan oleh Moro et al. (2014).

## Feature Scaling: Alasan Penggunaan StandardScaler



### Alasan Utama Memilih StandardScaler

`StandardScaler` melakukan transformasi dengan rumus, sehingga setiap fitur memiliki:
- Mean ≈ 0
- Standard deviation ≈ 1

Pemilihan StandardScaler didasarkan pada beberapa pertimbangan berikut:

---

#### 1. Sesuai dengan Asumsi Logistic Regression
Logistic Regression adalah **model linear** yang:
- Sensitif terhadap skala fitur
- Menggunakan optimasi berbasis **gradient descent**
- Mengasumsikan distribusi fitur relatif terpusat di sekitar mean

StandardScaler sangat cocok karena:
- Menjaga hubungan linear antar fitur
- Tidak memampatkan distribusi seperti MinMaxScaler
- Membuat proses konvergensi lebih stabil dan cepat

---

#### 2. Interpretasi Koefisien Menjadi Lebih Bermakna
Dengan StandardScaler:
- Setiap koefisien merepresentasikan perubahan log-odds akibat **kenaikan 1 standar deviasi** fitur
- Perbandingan pengaruh antar fitur menjadi lebih adil
- Analisis feature importance berbasis koefisien menjadi lebih valid

Hal ini penting karena Logistic Regression digunakan tidak hanya untuk prediksi, tetapi juga **interpretasi bisnis**.

---

#### 3. Lebih Stabil terhadap Outlier dibanding MinMaxScaler
Dataset perbankan memiliki potensi outlier, terutama pada fitur seperti:
- `balance`
- `duration`

StandardScaler:
- Tidak memaksa semua nilai berada di rentang [0,1]
- Lebih stabil terhadap nilai ekstrem dibanding MinMaxScaler
- Tetap mempertahankan bentuk distribusi data

---

### Mengapa Tidak Menggunakan Scaler Lain?

#### MinMaxScaler
Tidak dipilih karena:
- Sangat sensitif terhadap outlier
- Memampatkan distribusi data
- Kurang ideal untuk model linear berbasis gradien

#### RobustScaler
Tidak dipilih karena:
- Lebih cocok jika mayoritas fitur mengandung outlier ekstrem
- Dalam kasus ini, outlier ditangani melalui preprocessing lain (log transform / handling khusus)

#### Tidak Melakukan Scaling
Tidak memungkinkan karena:
- Logistic Regression sangat sensitif terhadap perbedaan skala fitur
- Dapat menyebabkan model bias ke fitur berskala besar

---

### Kesimpulan
StandardScaler dipilih karena:
- Paling sesuai dengan karakteristik **Logistic Regression**
- Menstabilkan proses training dan optimasi
- Mempermudah interpretasi koefisien
- Memberikan keseimbangan antara stabilitas dan informasi distribusi data

Dengan demikian, penggunaan StandardScaler merupakan keputusan yang **tepat secara teknis dan relevan secara bisnis** dalam konteks pemodelan telemarketing deposit ini.

## 7.1 Numeric Feature Pipeline

````python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
````

1. Handling nilai numerik ekstrem dan nilai khusus:
   Beberapa fitur numerik seperti `pdays` dan `previous` mengandung nilai khusus 
   (misalnya -1 atau 999) yang merepresentasikan kondisi operasional 
   "nasabah belum pernah dihubungi", bukan missing value murni.

   Pada tahap ini, nilai tersebut **tidak dihapus atau diganti**, karena:
   - Nilai tersebut membawa informasi perilaku historis nasabah
   - Menghapusnya justru menghilangkan sinyal penting terkait peluang konversi
   - Dalam konteks bisnis, status "belum pernah dihubungi" adalah kondisi valid

   Oleh karena itu, nilai tersebut dibiarkan apa adanya dan diperlakukan 
   sebagai bagian dari distribusi numerik yang sah.

2. Median Imputation:
   Median dipilih karena lebih robust terhadap outlier, terutama pada variabel 
   seperti `balance` yang memiliki distribusi skewed dan rentang nilai lebar.

3. StandardScaler:
   StandardScaler digunakan karena:
   - Model yang digunakan (Logistic Regression dan model berbasis optimisasi) 
     sensitif terhadap skala fitur
   - Fitur numerik memiliki rentang yang tidak bounded dan sangat bervariasi
   - Standardisasi membantu proses konvergensi optimisasi menjadi lebih stabil

   Penggunaan StandardScaler bukan keputusan default, tetapi didasarkan pada 
   karakteristik model dan distribusi fitur numerik.


## 7.2 Categorical Feature Pipeline

````python
from sklearn.preprocessing import OneHotEncoder

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])
````

### Imputasi Kategorikal

Missing value pada fitur kategorikal diisi menggunakan **nilai modus** (kategori
yang paling sering muncul). Pendekatan ini dipilih untuk menjaga **konsistensi
distribusi kategori** pada data tanpa menambahkan kategori baru yang bersifat
artifisial.

Penggunaan modus sebagai metode imputasi memiliki beberapa keunggulan:
- Mempertahankan karakteristik mayoritas data
- Menghindari fragmentasi kategori yang dapat meningkatkan sparsity
- Sederhana, stabil, dan mudah diinterpretasikan
- Minim asumsi tambahan terhadap struktur data

Pendekatan ini sangat sesuai untuk konteks data telemarketing, di mana kategori
mayoritas merepresentasikan pola perilaku nasabah yang paling umum dan relevan
secara bisnis.

---

### One-Hot Encoding

One-Hot Encoding digunakan untuk mentransformasi fitur kategorikal menjadi
representasi numerik karena sebagian besar fitur kategorikal dalam dataset
tidak memiliki hubungan ordinal alami, seperti:
- `job`
- `marital`
- `education`
- `contact`
- `month`

Penerapan ordinal encoding pada fitur-fitur tersebut berisiko menciptakan
**urutan numerik semu** yang tidak memiliki makna konseptual dan dapat
menyesatkan model, terutama model berbasis linear dan jarak.

Selain itu, target encoding tidak digunakan karena:
- Berpotensi menyebabkan **data leakage** apabila tidak diimplementasikan
  dengan sangat hati-hati
- Risiko overfitting meningkat pada dataset dengan **class imbalance**
- Nilai encoding dapat secara implisit membawa informasi target ke data training

Dalam konteks baseline dan evaluasi model, One-Hot Encoding dipilih karena:
- Tidak menyisipkan asumsi hubungan antar kategori
- Aman terhadap data leakage
- Stabil terhadap perubahan distribusi data
- Mudah diaudit dan dijelaskan kepada stakeholder non-teknis

Dengan pertimbangan tersebut, One-Hot Encoding menjadi pendekatan yang paling
sesuai untuk membangun pipeline preprocessing yang transparan, robust, dan
reproducible pada kasus prediksi keberhasilan telemarketing.

## 8.4 Cross-Validation (Exploratory Benchmark)

Cross-validation dilakukan **hanya untuk eksplorasi awal**, bukan sebagai dasar keputusan akhir.

````python
from sklearn.model_selection import StratifiedKFold, cross_val_score
import pandas as pd

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = []

for name, model in models.items():
    estimator = Pipeline([
        ('preprocess', preprocessor),
        ('model', model)
    ])
    
    scores = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv=cv,
        scoring='precision'
    )
    
    cv_results.append({
        "Model": name,
        "Mean Precision (CV)": scores.mean(),
        "Std Precision (CV)": scores.std()
    })

cv_results_df = pd.DataFrame(cv_results).sort_values(
    by="Mean Precision (CV)", ascending=False
)

cv_results_df
````

Stratified K-Fold Cross-Validation digunakan untuk mengevaluasi kestabilan 
precision pada berbagai subset data training, dengan menjaga proporsi kelas 
positif dan negatif di setiap fold.

Cross-validation pada tahap ini bersifat eksploratif dan bertujuan untuk:
1. Mengidentifikasi model yang memiliki performa precision yang konsisten
2. Menghindari ketergantungan pada satu pembagian data (single split bias)
3. Memberikan gambaran awal model mana yang layak dievaluasi lebih lanjut

Hasil cross-validation TIDAK digunakan sebagai dasar pemilihan model final.  
Keputusan akhir tetap dilakukan menggunakan hold-out test set agar 
merepresentasikan performa pada data yang benar-benar belum pernah dilihat model.

### 9.3.1 Pertimbangan Metodologis: Synthetic Minority Over-sampling Technique (SMOTE)

**Synthetic Minority Over-sampling Technique (SMOTE)** merupakan metode penyeimbangan data (*resampling*) yang digunakan untuk mengatasi masalah **ketidakseimbangan kelas (class imbalance)**. Berbeda dengan metode *random oversampling* yang hanya menggandakan data minoritas, SMOTE menghasilkan **sampel sintetis baru** melalui proses **interpolasi linier antar observasi kelas minoritas yang berdekatan di ruang fitur**.

Secara konseptual, SMOTE dibangun di atas beberapa asumsi utama, yaitu:
- Fitur bersifat **numerik dan kontinu**
- Jarak antar observasi dalam ruang fitur memiliki **makna geometris**
- Interpolasi antar titik data dapat merepresentasikan **observasi baru yang realistis**

#### Tujuan Penggunaan SMOTE
Tujuan utama penggunaan SMOTE adalah:
1. **Meningkatkan representasi kelas minoritas** agar model tidak bias terhadap kelas mayoritas  
2. **Membantu model pembelajaran mesin mempelajari pola kelas minoritas** secara lebih efektif  
3. **Mengurangi risiko model terjebak pada akurasi semu (*accuracy trap*)**, di mana model tampak akurat tetapi gagal mengenali kelas minoritas  

Pada dataset dengan fitur numerik murni dan struktur kontinu, SMOTE sering terbukti mampu meningkatkan metrik seperti *recall*, *F1-score*, dan keseimbangan performa antar kelas.

#### Kondisi Ideal Penggunaan SMOTE
SMOTE umumnya **layak digunakan** dalam kondisi berikut:
- Dataset didominasi oleh **fitur numerik kontinu**
- Hubungan antar fitur memiliki **interpretasi geometris yang jelas**
- Data tidak banyak mengandung variabel kategorikal diskrit
- Tujuan utama adalah **optimasi performa prediktif**
- Risiko distorsi semantik data relatif kecil

Dalam konteks tersebut, interpolasi antar titik data minoritas masih dapat menghasilkan sampel sintetis yang masuk akal dan representatif.

#### Keterbatasan SMOTE pada Kasus Telemarketing
Pada kasus **telemarketing untuk prediksi langganan deposito berjangka (term deposit subscriptions)**, karakteristik data **tidak sepenuhnya memenuhi asumsi dasar SMOTE**, dengan beberapa pertimbangan metodologis utama sebagai berikut:

- Mayoritas fitur bersifat **kategorikal**, antara lain:
  - pekerjaan (*job*)
  - status pernikahan (*marital*)
  - tingkat pendidikan (*education*)
  - jenis kontak (*contact*)
- Fitur-fitur tersebut direpresentasikan menggunakan **One-Hot Encoding**
- Dataset bersifat **tabular heterogen**, mencerminkan profil dan perilaku nasabah yang diskrit

Dalam kondisi ini, proses interpolasi SMOTE pada fitur hasil One-Hot Encoding berpotensi menghasilkan **nilai sintetis yang tidak bermakna secara semantik**, seperti:
- Kombinasi parsial dari beberapa kategori sekaligus
- Profil nasabah sintetis yang **tidak mungkin terjadi dalam interaksi telemarketing nyata**
- Distorsi distribusi kategori yang memiliki arti penting secara bisnis

Dengan kata lain, meskipun secara matematis memungkinkan, sampel hasil SMOTE **tidak mencerminkan perilaku nasabah yang realistis dalam konteks kampanye telemarketing**.

#### Implikasi terhadap Interpretasi dan Validitas Model
Penerapan SMOTE pada konteks ini berpotensi menimbulkan beberapa risiko, antara lain:
- **Synthetic samples tidak merepresentasikan keputusan nasabah yang nyata**
- **Distribusi data menjadi tidak alami dari sudut pandang operasional bisnis**
- Interpretasi model menjadi **kurang kredibel bagi pengambil keputusan**
- Meningkatnya risiko *overfitting* terhadap pola sintetis yang tidak realistis

Dalam analisis **perilaku nasabah dan efektivitas kampanye telemarketing**, keandalan data dan interpretabilitas model menjadi aspek yang lebih krusial dibandingkan sekadar peningkatan metrik evaluasi.

#### Peran SMOTE dalam Analisis Telemarketing Term Deposit
Berdasarkan pertimbangan metodologis dan konteks bisnis telemarketing, **SMOTE digunakan secara terbatas sebagai alat eksploratif**, bukan sebagai pendekatan utama dalam pemodelan prediksi langganan deposito berjangka.

## 10.4 Proses Hyperparameter Tuning (GridSearchCV)

````python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# Stratified CV untuk menjaga proporsi kelas
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

# Pipeline final untuk tuning
tuning_pipeline = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('model', LogisticRegression(random_state=42))
])

# Ruang hyperparameter (terkontrol & tidak eksploratif)
param_grid = {
    'model__C': [0.01, 0.1, 1.0, 10],
    'model__penalty': ['l2'],
    'model__solver': ['lbfgs']
}

# GridSearch dengan precision sebagai metrik utama
grid_search = GridSearchCV(
    estimator=tuning_pipeline,
    param_grid=param_grid,
    scoring='precision',
    cv=cv,
    n_jobs=-1,
    verbose=1
)

# Fit hanya pada data training
grid_search.fit(X_train, y_train)
````

````python
# Pisahkan kolom numerik & kategorikal
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns

# Imputasi numerik → median
X[num_cols] = X[num_cols].fillna(X[num_cols].median())

# Imputasi kategorikal → "Unknown"
X[cat_cols] = X[cat_cols].fillna("Unknown")
````

# 11. Model Interpretation

## 11.1 Tujuan Model

Model machine learning pada proyek ini dibangun untuk memprediksi apakah
seorang nasabah akan melakukan *subscription* terhadap produk deposito
berjangka (`y = yes`) berdasarkan data hasil kampanye telemarketing.

Tujuan utama model **bukan sekadar memaksimalkan jumlah prediksi positif**,
melainkan **memastikan bahwa nasabah yang diprediksi “yes” memiliki peluang
tinggi untuk benar-benar melakukan subscription**, sehingga biaya dan effort
telemarketing dapat digunakan secara efisien.

Oleh karena itu, **precision dipilih sebagai metrik utama** dalam evaluasi dan
pemilihan model.

---

## 11.2 Konteks Bisnis dan Implikasi Kesalahan Prediksi

Dalam konteks telemarketing perbankan:

- **False Positive (FP)**  
  Model memprediksi nasabah akan berlangganan, namun kenyataannya tidak.
  → Mengakibatkan pemborosan biaya telepon, waktu agen, dan potensi penurunan
  efektivitas kampanye.

- **False Negative (FN)**  
  Model memprediksi nasabah tidak akan berlangganan, padahal sebenarnya
  berpotensi.
  → Kehilangan peluang, namun dampaknya relatif lebih kecil dibanding FP karena
  sumber daya tetap dapat dialokasikan ke nasabah lain dengan probabilitas lebih
  tinggi.

Karena **biaya FP lebih signifikan daripada FN**, precision menjadi metrik yang
paling selaras dengan tujuan bisnis.

---

## 11.3 Interpretasi Pipeline dan Konsistensi Transformasi

Seluruh proses preprocessing dan modeling dibungkus dalam satu `Pipeline`
dengan `ColumnTransformer`, yang mencakup:

- Imputasi nilai hilang pada fitur numerik dan kategorikal
- One-Hot Encoding untuk fitur kategorikal
- Feature scaling menggunakan `StandardScaler` untuk fitur numerik
- Model klasifikasi sebagai estimator akhir

Pendekatan ini memastikan bahwa:
- Transformasi hanya di-*fit* pada data training
- Data testing dan data inferensi diproses secara identik
- Tidak terjadi data leakage
- Interpretasi performa model mencerminkan kondisi dunia nyata

---

## 11.4 Interpretasi Evaluasi Model

Evaluasi model dilakukan pada **test set** menggunakan beberapa metrik:

- **Precision (metrik utama)**  
  Mengukur seberapa besar proporsi prediksi “yes” yang benar-benar tepat.
  Nilai precision yang tinggi menunjukkan bahwa daftar nasabah hasil seleksi
  model memiliki kualitas tinggi dan layak ditindaklanjuti oleh tim
  telemarketing.

- **Recall (metrik pendukung)**  
  Digunakan untuk memahami sejauh mana model mampu menangkap seluruh potensi
  nasabah yang benar-benar akan berlangganan. Recall tidak dijadikan dasar
  keputusan utama karena fokus proyek adalah efisiensi, bukan cakupan maksimal.

- **F1-Score (metrik pendukung)**  
  Memberikan gambaran keseimbangan antara precision dan recall, sehingga dapat
  digunakan untuk menilai stabilitas performa model secara umum.

- **Accuracy (metrik informatif)**  
  Ditampilkan sebagai referensi global, namun tidak menjadi indikator utama
  karena dataset bersifat imbalanced dan accuracy dapat memberikan gambaran
  yang menyesatkan.

**ROC-AUC tidak digunakan**, karena:
- Bersifat threshold-independent
- Tidak merepresentasikan keputusan operasional nyata
- Kurang relevan untuk skenario seleksi nasabah berbasis threshold tertentu

   Integrasikan bobot biaya kesalahan klasifikasi secara eksplisit ke dalam proses training untuk merepresentasikan perbedaan dampak antara *false positive* dan *false negative* sesuai dengan prioritas bisnis.

2. **Monitoring Data Drift dan Retraining Berkala**  
   Lakukan pemantauan distribusi fitur dan performa model secara periodik. Jika terjadi penurunan precision atau perubahan pola perilaku nasabah yang signifikan, model perlu dilatih ulang menggunakan data terbaru.

3. **Eksperimen dan Validasi Berbasis A/B Testing**  
   Lakukan eksperimen terkontrol antara kampanye berbasis model dan kampanye konvensional untuk mengukur dampak langsung terhadap conversion rate dan Return on Investment (ROI) pemasaran.

---

### 13.2.4 Rekomendasi Pengembangan Sistem dan Teknologi

1. **Integrasi dengan Sistem CRM dan Call Center**  
   Integrasikan model ke dalam sistem CRM atau dashboard operasional sehingga agen dapat mengakses skor probabilitas nasabah secara real-time dan memperoleh daftar prioritas kontak harian secara otomatis.

2. **Pengembangan Aplikasi Internal untuk Monitoring dan Kontrol**  
   Kembangkan antarmuka berbasis web (misalnya menggunakan Streamlit atau framework serupa) yang memungkinkan tim marketing dan supervisor untuk mengatur threshold, memantau performa kampanye, serta melakukan analisis hasil secara mandiri.

## 13.3 Final Remarks

Proyek ini menegaskan bahwa keberhasilan implementasi machine learning tidak hanya ditentukan oleh kinerja teknis model, tetapi oleh keselarasan antara tujuan bisnis, pemilihan metrik evaluasi, serta integrasi model ke dalam proses operasional organisasi.

Dengan pendekatan yang berfokus pada precision, analisis trade-off yang disadari, dan penerjemahan hasil model ke dalam rencana aksi yang terstruktur, solusi yang dikembangkan memberikan dasar yang kuat bagi pengambilan keputusan telemarketing yang lebih efisien, terukur, dan berkelanjutan, sekaligus membuka peluang pengembangan lanjutan menuju sistem pemasaran berbasis data yang lebih matang dan adaptif.


# 14. Streamlit Application & Tableau Business Dashboard – Decision Support System

Sebagai implementasi akhir project, dikembangkan dua komponen utama berbasis visual analytics, yaitu **aplikasi Streamlit** dan **dashboard Tableau**, yang secara bersama-sama membentuk sebuah **Decision Support System (DSS)** untuk mendukung optimalisasi campaign telemarketing produk **Bank Term Deposit**.

Kedua platform memiliki peran yang saling melengkapi:
- **Streamlit** digunakan untuk *operational-level decision making* (scoring nasabah individual)
- **Tableau** digunakan untuk *strategic-level analysis* dan pemantauan performa campaign secara agregat

---

## 14.1 Streamlit Application – Customer Scoring & Operational Decision Support

Aplikasi **Streamlit** berfungsi sebagai **alat scoring nasabah** yang mengintegrasikan model machine learning ke dalam antarmuka interaktif yang mudah digunakan oleh stakeholder bisnis.

### Fungsi Aplikasi
Aplikasi menghasilkan **probabilitas berlangganan deposito** pada level individu nasabah, yang kemudian digunakan untuk melakukan **segmentasi prioritas telemarketing**, yaitu:
- **Low Priority**
- **Medium Priority**
- **High Priority**

Pendekatan ini memungkinkan tim telemarketing memfokuskan upaya pada nasabah dengan potensi konversi tertinggi.

### Target Pengguna
Aplikasi ditujukan untuk:
- Manajer Marketing  
- Tim Sales / Telemarketing  

Pengguna tidak memerlukan latar belakang teknis atau pemahaman machine learning untuk mengoperasikan aplikasi.

### Skenario Penggunaan
Pengguna menginput data karakteristik nasabah melalui sidebar aplikasi. Sistem kemudian:
1. Memproses data menggunakan pipeline machine learning
2. Menghasilkan probabilitas konversi
3. Menampilkan segmentasi prioritas nasabah
4. Memberikan rekomendasi tindakan bisnis

Aplikasi dapat digunakan baik sebagai **simulasi sebelum campaign** maupun sebagai **alat bantu keputusan operasional** saat campaign berlangsung.

### Keputusan Bisnis yang Didukung
Aplikasi Streamlit membantu:
- Menentukan nasabah yang perlu dihubungi terlebih dahulu  
- Mengoptimalkan alokasi resource telemarketing  
- Mengurangi biaya akibat panggilan yang tidak efektif  
- Meningkatkan conversion rate campaign
